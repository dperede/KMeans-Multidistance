```python
from KMeans_Class import KMeans
from sklearn import datasets
%matplotlib inline
```

```python
# We load the iris dataset to use it in this example. We take the save the data and the target in two different variables
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Refer to the documentation to see the possibilities of the parameters for the constructor
model = KMeans(k=5, maxiter =300, distance='cosine', record_heterogeneity=[], verbose=True, seed=123)
```

The library follows the sklearn API and it is only required to call the fit method to run the algorithm

```python
centroids, cluster_assignment = model.fit(X)
```

The output is the centroids of the clusters that have been calculated and the cluster assignments of all the data points that have been used to fit the model

```python
centroids
```

```python
cluster_assignment
```

If the parameter record_heterogeneity has been provided in the constructor, the heterogeneity calculated at each iteration of the process to fit the model can be printed calling to the plot_heterogeneity function

```python
model.plot_heterogeneity()
```
